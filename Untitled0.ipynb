```python
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
df = sns.load_dataset('titanic')
```

```python
df = df[['age', 'fare', 'sex', 'embarked', 'class', 'survived']].dropna()
```

```python
df_encoded = pd.get_dummies(df, columns=['sex', 'embarked', 'class'], drop_first=True)
```

```python
X = df_encoded.drop('survived', axis=1)
Y = df_encoded['survived']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=True)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
print("✅ Data Shapes:")
print("X_train:", X_train_scaled.shape)
print("X_test:", X_test_scaled.shape)
print("Y_train:", Y_train.shape)
print("Y_test:", Y_test.shape)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
X_train_df = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_train_df['survived'] = Y_train.values
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=X_train_df, x='age', y='fare', hue='survived', palette='Set1')
plt.title('Age vs Fare (Scaled) Colored by Survival')
plt.xlabel('Age (Standardized)')
plt.ylabel('Fare (Standardized)')
plt.grid(True)
plt.show()
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import pandas as pd
```

```python
iris = load_iris()
X = iris.data
Y = iris.target
```

```python
df = pd.DataFrame(X, columns=iris.feature_names)
df['species'] = iris.target
print(df.head())
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, Y_train)
```

```python
Y_pred = knn.predict(X_test)
```

```python
print("\nAccuracy:", accuracy_score, Y_pred)
```

```python
print("\nConfution Matrix:\n", confusion_matrix(Y_test, Y_pred))
```

```python
print("\nClasification Report:\n", classification_report(Y_test, Y_pred, target_names=iris.target_names))
```
